```python
# Lets import a data from this library "chickwts"

from pydataset import data

df = data('chickwts')

df
```

PS 2: Total of 71 chickens are feed with different foods. The chicken company want to know which feed is the best to produce heavy chickens so that you can make more money.

```python
# Lets find the average weight of each food type

df.groupby(['feed'])['weight'].mean()
```

 if there is a significante difference in the average weight of each food type or not

Null Statement: No Significante difference in the average weight of each food Alternative: There is a significant difference in the average weight of each food

Which test: ANOVA as the data points are greater than 30 and the data is divided into groups.

For ANOVA, you need to use a library scipy

```python
# Lets seprate the groups

casein_group = df[df['feed'] == "casein"]
horsebean_group = df[df['feed'] == "horsebean"]
linseed_group = df[df['feed'] == "linseed"]
meatmeal_group = df[df['feed'] == "meatmeal"]
soybean_group = df[df['feed'] == "soybean"]
sunflower_group = df[df['feed'] == "sunflower"]
```

```python
# Get the library

from scipy.stats import stats

# Syntax - stats.f_oneway(Group1, Group2, Group3,....)

stats.f_oneway(casein_group['weight'], horsebean_group['weight'], linseed_group['weight'],
              meatmeal_group['weight'], soybean_group['weight'], sunflower_group['weight'])
```

p-value ~ 0.00 which is less than 0.05 thus, reject my null statement, thus, we will go with alternative statement i.e There is a significant difference in the average weight of each food. Thus Sunflower is the best feed to produce heavy chickens.
